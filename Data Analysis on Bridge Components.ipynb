### Exploring Yearbuilt w.r.t Subnumber Interventions

```python
#Importing required libraries
import pandas as pd
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')

#Reading the data
df = pd.read_csv('nebraska_deep.csv')
#print(df.columns)

#Removing the duplicates
df.drop_duplicates(subset=['structureNumber'], keep='last', inplace=True)

#Replcaing the int values to strings
df['subNumberIntervention']=df['subNumberIntervention'].replace([0.0,1.0,2.0,3.0], ['Zero','One','One', 'One'])

#Removing the outliers
df2 = df[df.yearBuilt >= 500]
df2 = df2[df2.yearBuilt != 1935]

#Subsetting and normalizing the dataframe 
df3=df2.groupby(['yearBuilt','subNumberIntervention'])['yearBuilt'].count().to_frame()
df3 = df3.rename({'yearBuilt': 'count'}, axis=1)
df3=df3.reset_index()
df7=df3
df7['normalization']=(df7['count']-df7['count'].min()) / (df7['count'].max()-df7['count'].min())
df4 = df3[df3.subNumberIntervention != 'One']
df4['normalization']=(df4['count']-df4['count'].min()) / (df4['count'].max()-df4['count'].min())
df5=df3[df3.subNumberIntervention != 'Zero']
df5['normalization']=(df5['count']-df5['count'].min()) / (df5['count'].max()-df5['count'].min())
df6=df4.append(df5, ignore_index=True)

#Plotting
fig = px.bar(df7, x='yearBuilt',y='normalization',color='subNumberIntervention',barmode="group" )
fig.show()
```
